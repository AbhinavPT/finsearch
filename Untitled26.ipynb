```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load the data from the CSV file
data = pd.read_csv('option_data.csv')

# Constants
S = 24572.65  # Current stock price
r = 0.06      # Risk-free interest rate (annualized)
T = 3 / 365   # Time to expiry (in years)

# Black-Scholes Formula
def black_scholes_call_price(S, K, r, T, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# Apply Black-Scholes formula
data['Calculated_Price'] = data.apply(lambda row: black_scholes_call_price(S, row['STRIKE'], r, T, row['IV'] / 100), axis=1)

# Calculate errors
data['Price_Diff'] = data['LTP'] - data['Calculated_Price']
mae = mean_absolute_error(data['LTP'], data['Calculated_Price'])
mse = mean_squared_error(data['LTP'], data['Calculated_Price'])
rmse = np.sqrt(mse)

# Calculate percentage metrics
mape = (np.abs(data['Price_Diff']) / data['LTP']).mean() * 100
avg_market_price = data['LTP'].mean()
percentage_rmse = (rmse / avg_market_price) * 100

# Print accuracy metrics
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
print(f"Percentage RMSE: {percentage_rmse:.2f}%")

# Plotting
plt.figure(figsize=(12, 6))

# Market Price vs Calculated Price
plt.subplot(1, 2, 1)
plt.plot(data['STRIKE'], data['LTP'], 'o-', label='Market Price')
plt.plot(data['STRIKE'], data['Calculated_Price'], 's--', label='Calculated Price')
plt.xlabel('Strike Price')
plt.ylabel('Price')
plt.title('Market Price vs Black-Scholes Calculated Price')
plt.legend()
plt.grid(True)

# Price Difference
plt.subplot(1, 2, 2)
plt.plot(data['STRIKE'], data['Price_Diff'], 'o-', color='red')
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('Strike Price')
plt.ylabel('Price Difference')
plt.title('Price Difference (Market - Calculated)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

# Constants
S = 24572.65  # Current stock price
r = 0.06      # Risk-free interest rate (annualized)
T = 3 / 365   # Time to expiry (in years)

# Black-Scholes Formula for Call Options
def black_scholes_call_price(S, K, r, T, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# Black-Scholes Formula for Put Options
def black_scholes_put_price(S, K, r, T, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    put_price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    return put_price

# Define data for call options
call_data = {
    'STRIKE': [24350.00, 24400.00, 24450.00, 24500.00, 24550.00, 24600.00, 24650.00, 24700.00, 24750.00, 24800.00, 24850.00, 24900.00],
    'LTP': [252.85, 216.40, 177.30, 144.50, 117.00, 90.85, 69.00, 51.00, 35.80, 25.00, 16.95, 11.45],
    'IV': [7.14, 8.77, 9.43, 9.80, 10.12, 10.27, 10.42, 10.46, 10.42, 10.49, 10.51, 10.57]
}

# Define data for put options
put_data = {
    'STRIKE': [24350.00, 24400.00, 24450.00, 24500.00, 24550.00, 24600.00, 24650.00, 24700.00, 24750.00, 24800.00, 24850.00, 24900.00],
    'LTP': [37.4, 48.4, 62, 79, 98.7, 124.35, 151, 180, 218.6, 257.3, 299.15, 343.75],
    'IV': [14.19, 14.06, 14, 13.86, 13.96, 14.04, 14.2, 14.43, 14.81, 15.38, 15.86, 16.66]
}

# Create DataFrames
df_calls = pd.DataFrame(call_data)
df_puts = pd.DataFrame(put_data)

# Calculate Black-Scholes prices
df_calls['Calculated_Price'] = df_calls.apply(lambda row: black_scholes_call_price(S, row['STRIKE'], r, T, row['IV'] / 100), axis=1)
df_puts['Calculated_Price'] = df_puts.apply(lambda row: black_scholes_put_price(S, row['STRIKE'], r, T, row['IV'] / 100), axis=1)

# Add a 'Type' column to distinguish call and put options
df_calls['Type'] = 'Call'
df_puts['Type'] = 'Put'

# Combine call and put data into a single DataFrame
combined_df = pd.concat([df_calls, df_puts], ignore_index=True)

# Reorder columns
combined_df = combined_df[['Type', 'STRIKE', 'LTP', 'IV', 'Calculated_Price']]

# Display the table
import matplotlib.pyplot as plt
import seaborn as sns

# Display DataFrame as a table
plt.figure(figsize=(12, 8))
sns.heatmap(combined_df[['Type', 'STRIKE', 'LTP', 'IV', 'Calculated_Price']].set_index(['Type', 'STRIKE']), annot=True, fmt=".2f", cmap='coolwarm', cbar=False)
plt.title('Call and Put Options: Market Price vs Black-Scholes Calculated Price')
plt.show()
```
